```python
import re

input_file = "maithili_corpus.txt"
output_file = "maithili_cleaned.txt"
max_sentences = 30000

all_sentences = []

with open(input_file, 'r', encoding='utf-8') as f:
    for line in f:
        line = line.strip()
        if not line:
            continue

        #Split line into sentences by Devanagari danda
        for sentence in re.split(r'[।॥]', line):
            s = sentence.strip()          #remove spaces
            s = s.rstrip('|')             #remove trailing |
            #Remove English letters, English digits, and Devanagari digits
            s = re.sub(r'[A-Za-z0-9\u0966-\u096F]', '', s)
            
            if s:
                all_sentences.append(s) 
                if len(all_sentences) >= max_sentences:
                    break
        if len(all_sentences) >= max_sentences:
            break

with open(output_file, 'w', encoding='utf-8') as f:
    for s in all_sentences:
        f.write(s + "\n")
print(f"Done! Total sentences: {len(all_sentences)}")
print(f"Saved to: {output_file}")
```

```python
import re

input_file = "maithili_corpus.txt"
output_file = "maithili_cleaned1.txt"

with open(input_file, 'r', encoding='utf-8') as f:
    text = f.read()

# Remove English letters and digits
text = re.sub(r'[A-Za-z0-9]', '', text)

# Remove Devanagari digits (०-९)
text = re.sub(r'[\u0966-\u096F]', '', text)

# Remove extra spaces and empty lines
text = re.sub(r'\s+', ' ', text).strip()

with open(output_file, 'w', encoding='utf-8') as f:
    f.write(text)

print("Cleaning complete!")
print("Saved to:", output_file) 
```

```python
#Downsampling data
import random
import re

target_size = 179000  

lang_file = "maithili_cleaned1.txt"
with open(lang_file, "r", encoding="utf-8") as f:
    words = f.read().split()

#Clean words: keep only Devanagari
def clean_word(word):
    word = re.sub(r'[^ऀ-ॿ]', '', word)  #remove non-Devanagari
    return word.strip()

#Clean all words
cleaned_words = [clean_word(w) for w in words if clean_word(w)]

#Downsample to exact target size
if len(cleaned_words) > target_size:
    cleaned_words = random.sample(cleaned_words, target_size)
elif len(cleaned_words) < target_size:
    
    cleaned_words = cleaned_words * (target_size // len(cleaned_words)) + cleaned_words[:target_size % len(cleaned_words)]

#Save
out_file = "maithili_balanced.txt"
with open(out_file, "w", encoding="utf-8") as f:
    f.write(" ".join(cleaned_words))

print(f"Maithili done: {len(cleaned_words)} words saved as {out_file}")
```
